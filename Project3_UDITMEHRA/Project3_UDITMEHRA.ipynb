```python
# a) Add a new column “TotalAdvt” to the dataframe df_marketing, which will have the total of all the three marketing
# expenditures (tv, radio and news-paper)
df_marketing['TotalAdvt']=df_marketing['TV']+df_marketing['Radio']+df_marketing['Newspaper']
df_marketing
```

```python
# b) Fit a Simple Linear Regression for Sales and  TotalAdvt and find out the values of intercept slope and the R-square value.
#    Describe the model in your terms. The model should be named simple_model.
a=df_marketing['TotalAdvt']
x=a.values.reshape(-1,1)
y=df_marketing['Sales']
```

```python
from sklearn import datasets, linear_model
simple_model = linear_model.LinearRegression()
simple_model.fit(x,y)
```

```python
# find out the values of intercept slope
simple_model.intercept_
```

```python
simple_model.coef_
```

```python
# R-square value
simple_model.score(x,y)
```

```python
plt.scatter(x,y, color='red')
plt.plot(x,simple_model.predict(x),color='blue', linewidth=3)
```

# Describe the model in your terms
Simple linear regression is an approach for predicting a quantitative response using a single feature (or "predictor" or "input variable") It takes the following form: y=β0+β1x where, y is the response x is the feature β0 is the intercept β1 is the coefficient for x β0 and β1 are called the model coefficients.

As we know that Sales is dependent variables and tv,radio,newspaper are predictor variable. In above model y =Sales and x=tv,radio,newspaper. And we have determined the intercept ,R-squared value and coefficients using linear regression.

```python
# c) What will be the Sales using the above Model if total expenditure
#    on advertisements is Rupees 50,000.Show the steps done to find out the Sales value. 
sales= simple_model.intercept_ + simple_model.coef_*50
sales
```

# Equation used:
y=β0+β1x
β1 - This is the SLOPE of the regression line. Thus this is the amount that the Y variable (dependent) will change for each 1 unit change in the X variable.
β0 - This is the intercept of the regression line with the y-axis. In otherwords it is the value of Y if the value of X = 0.
y= 4.129453401328467 + 0.04978274×50
We would use 50 instead of 50,000 because the original data consists of examples that are divided by 1000

```python
# d) Perform part c) using the predict method for the model created in part b)
simple_model.predict(50)
```

```python
# a) Fit a Simple Linear Regression for Sales while keeping tv,
#    radio and newspaper as three different x-parameters. The model should be named multi_model.
from sklearn.linear_model import SGDRegressor,LinearRegression
TV= df_marketing['TV']
Radio= df_marketing['Radio']
Newspaper= df_marketing['Newspaper']
x1=np.vstack((TV,Radio,Newspaper))
x1=x1.T
y1=df_marketing['Sales']
```

```python
multi_model = linear_model.LinearRegression()
```

```python
multi_model.fit(x1,y1)
```

```python
# R-square value.
multi_model.score(x1,y1)
```

```python
multi_model.coef_
```

```python
# Find out the values of intercept slopes 
multi_model.intercept_
```

# Describe the model in your terms.
Simple linear regression can easily be extended to include multiple features. This is called multiple linear regression:

y=β0+β1x1+...+βnxn . Each x represents a different feature, and each feature has its own coefficient. In this case:

y= β0+β1×TV+β2×Radio+β3×Newspaper
We have fitted the Simple Linear Regression for Sales while keeping tv, radio and newspaper as three different x-parameters.

```python
# b) Write the Mathematical equation for Sales for this Model. Predict Sales for the below mentioned Expenditures
# i) tv 90000, radio 3000 and newspaper 45000

sale= multi_model.intercept_ + multi_model.coef_[0]*90 + multi_model.coef_[1]*3 + multi_model.coef_[2]*45
sale
```

```python
# ii) tv 290000 and newspaper 80000
sale= multi_model.intercept_ + multi_model.coef_[0]*290 + multi_model.coef_[2]*80
sale
```

```python
# c) Perform part b) using  the predict method for the model.
# i) tv 90000, radio 3000 and newspaper 45000
multi_model.predict([[90,3,45]])
```

```python
# ii) tv 290000 and newspaper 80000
multi_model.predict([[290,0,80]])
```

# d) Compare the result in part b) and c)
We can clearly see that results of both part b) and c) are same that means we have fitted the Simple Linear Regression for Sales while keeping tv, radio and newspaper as three different x-parameters correctly.
The calculated predict value of Sales is 7.677097348621903 when tv=90000, radio=3000 and newspaper=45000.
